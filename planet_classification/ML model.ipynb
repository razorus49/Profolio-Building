```python
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import os
import numpy as np
from skimage.feature import hog
from sklearn.naive_bayes import MultinomialNB
from joblib import dump, load
```

```python

def pad_features(features):
    max_length = max(len(arr) for arr in features)

    # Initialize a new array of zeros with the desired shape
    padded_features = np.zeros((len(features), max_length))

    # Copy the original arrays and pad with zeros
    for i, arr in enumerate(features):
        padded_features[i, :len(arr)] = arr

    return padded_features
```

```python
label_folder_path = r"labels"
feature_folder_path = r"features"
feature_test_folder_path = r"feature test folder"
label_test_folder_path = r"label test folder"

init_features_path = r"experimental data/features"
init_labels_path = r"experimental data/labels"
model = MultinomialNB()

features=[]
labels=[]
#initialize model with different classes
for file_name in os.listdir(init_features_path):
    feature_path = os.path.join(init_features_path, file_name)
    label_path = os.path.join(init_labels_path, file_name)

    feature = np.loadtxt(feature_path).flatten()
    label = np.loadtxt(label_path).tolist()
    features.append(feature.tolist())
    labels.append(label)

features = pad_features(features)
model.fit(features, labels)


j=0
for sub_folder in os.listdir(feature_folder_path):
    print(f"scanning {sub_folder}")
    
    i=0
    
    feature_subfolder_path = os.path.join(feature_folder_path, sub_folder)
    label_subfolder_path = os.path.join(label_folder_path, sub_folder)
    
    features=[]
    labels=[]
    
    
    for file_name in os.listdir(feature_subfolder_path):
        
        feature_path = os.path.join(feature_subfolder_path, file_name)
        label_path = os.path.join(label_subfolder_path, file_name)
        
        feature = np.loadtxt(feature_path).tolist()
        label = np.loadtxt(label_path).tolist()
        
        features.append(feature)
        labels.append(label)
        
        if i>=10:
            i=0
            features = pad_features(features)
            
            # features = np.asarray(features, dtype = object )
            # print(features.shape)
            # labels = np.asarray(labels)
            X_train, X_test, y_train, y_test = train_test_split(features, labels)

            # y_train = np.asarray(y_train).astype(int)
            model.fit(X_train, y_train)
            print("trained")

            features=[]
            labels=[]

            
            name = sub_folder + file_name
            f_test_path = os.path.join(feature_test_folder_path, name)
            l_test_path = os.path.join(label_test_folder_path, name)
            
            np.savetxt(f_test_path, X_test)
            np.savetxt(l_test_path, y_test)
        
        i+=1

```

```python
dump(model, 'model.joblib') 
```

```python
num=0
total=0

for file_name in os.listdir(feature_test_folder_path):
    f_test_path = os.path.join(feature_test_folder_path, file_name)
    l_test_path = os.path.join(label_test_folder_path, file_name)
    x = np.loadtxt(f_test_path)
    print(x.shape)
    y = np.loadtxt(l_test_path)

    y_pred = model.predict(x)
    total += accuracy_score(y, y_pred)
    num+=1
    print(total/num)
print(total/num)
```
